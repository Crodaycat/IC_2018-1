# MANEJANDO DOCUMENTOS EXCEL

### Apertura de archivos de excel

INCLUYENDO LATEX: $\alpha$

```python
import openpyxl #Importamos la libreria que si instalo con: $pip install openpyxl

doc_excel = openpyxl.load_workbook('address.xlsx') #Leyendo el archivo que esta en la misma carpeta de este programa
#El metodo load_workbook recibe un argumento tipo string que hace referencia a el path o ruta completa (No relativa) donde hemos almacenado nuestro archivo. Ej: 'C:\\Users\Carpeta\archivo.xls'
```

### Lectura de hojas de cálculo

```python
arreglo_nombre_pest = doc_excel.get_sheet_names() #Un arreglo de strings con las variables de las hojas de datos
print(arreglo_nombre_pest) #Imprimiento las hojas de datos
nombre_pest_sel = arreglo_nombre_pest[0]
print(nombre_pest_sel)

hoja = doc_excel.get_sheet_by_name(nombre_pest_sel) #Obteniendo la primer hoja
print(hoja)
#El metodo get_sheet_names nos devuelve una lista de strings, los cuales corresponden a todas las hojas disponibles en el archivo
title_hoja = hoja.title #Validando el nombre de la hoja obtenida
#mediante el metodo get_sheet_by_name y su argumento de tipo string, indicamos la hoja de calculo que vamos operar
print(title_hoja)
```

### Lectura de Celdas

```python
hoja.rows
#Obteniendo  todas las filas y columnas utilizadas o que tengan valor alguno

#Otra forma de ver tupla con las multiples filas y columnas de nuestra hoja, si deseamos verlo de otra forma, haciendo uso de un for, podemos recorrer fila por fila
for filas in hoja.rows:
    print(filas)
    
#Obteniendo por ejemplo  el valor A1 y de F2 de nuestra hoja y analizando que valores le da Python
    #Obteniendo según nombre de fila y columna de excel
print("A1 =", hoja['A1'].value)
print("El tipo de dato de A1 =", type(hoja['A1'].value))
print("D1 =", hoja['D1'].value)
print("El tipo de dato de D1 =", type(hoja['D1'].value))
print("F2 =", hoja['F2'].value)
print("El tipo de dato de F2 =", type(hoja['F2'].value))
    #Obteniendo segun que columna y fila en la coordenada (row,column) empezando ambas desde 1
print("Obteniendo el valor de A1, mediante coordenadas (row=1, column=1) =", hoja.cell(row=1,column=1).value) #Esto es lo mismo que A1
```

### Recorrer Hoja

```python
#recorrer los valores de nuestra hoja de forma básica (no es la mas optima):
for fila in hoja.rows:
    for columna in fila:
        print (columna.value)
        print ("")
```

### Recorrido de Filas y Columnas

```python
#otra forma de recorrido en la hoja de calculo, más optima
for fila in hoja.rows:
    for columna in fila:
        print(columna.coordinate,columna.value)
    print("----Final de Fila---")
```
